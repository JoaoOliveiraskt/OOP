```python
class PandasPublisher:
    def __init__(self):
        self.data = pd.DataFrame()

    def add_data(self, data_dict, column_name):
        if not isinstance(data_dict, dict):
            raise ValueError("Input data should be a dictionary.")

        new_data = pd.DataFrame(data_dict)
        if column_name in self.data:
            self.data[column_name] = new_data
        else:
            self.data = pd.concat([self.data, new_data], axis=1)

    def display_data(self):
        display(self.data)

    def save_to_excel(self, file_name):
        if self.data.empty:
            print("DataFrame is empty. Nothing to save.")
        else:
            self.data.to_excel(file_name, index=False)


diretorio = ('/content/drive/MyDrive/ColabNotebooks/POO')
publisher = PandasPublisher()

product1_data = {
    "Vendas": [100, 150, 200],
    "Custos": [50, 75, 90]
}

publisher.add_data(product1_data, "Produto1")

product2_data = {
    "Vendas": [120, 160, 180],
    "Custos": [60, 80, 95]
}

publisher.add_data(product2_data, "Produto2")

publisher.display_data()

publisher.save_to_excel(diretorio+"/dados_venda.xlsx")
```

## Uso de Class abstrata, herança e poliformismo


```python
class PandasBase:
    def __init__(self):
        self.data = pd.DataFrame()

    def add_data(self, data_dict):
        if not isinstance(data_dict, dict):
            raise ValueError("Input data should be a dictionary.")
        new_data = pd.DataFrame(data_dict)
        self.data = pd.concat([self.data, new_data], ignore_index=True)

    def display_data(self):
        display(self.data)

class PandasSubclass(PandasBase):
    def __init__(self, subclass_name):
        super().__init__()
        self.subclass_name = subclass_name

    def add_subclass_data(self, data_dict):
        if not isinstance(data_dict, dict):
            raise ValueError("Input data should be a dictionary.")
        new_data = pd.DataFrame(data_dict)
        new_data.columns = [f'{self.subclass_name}_{col}' for col in new_data.columns]
        self.data = pd.concat([self.data, new_data], axis=1)
```

```python
# Exemplo de uso:
base_publisher = PandasBase()
base_data = {'Nome': ['Alice', 'Bob'], 'Idade': [25, 30]}
base_publisher.add_data(base_data)
print("Base Publisher Data:")
base_publisher.display_data()

subclass_publisher = PandasSubclass("Subclass")
subclass_data = {'Salário': [50000, 60000]}
subclass_publisher.add_subclass_data(subclass_data)
print("\nSubclass Publisher Data:")
subclass_publisher.display_data()
```

```python
# Definindo uma interface (classe abstrata)
class IShape(ABC):
    @abstractmethod
    def area(self):
        pass

# Uma classe que implementa a interface IShape
class Circle(IShape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
```

```python
# Exemplo de uso
circle = Circle(5)
print(f"Área do círculo: {circle.area()}")
```

```python
class AbstractPandasPublisher(ABC):
    def __init__(self):
        self.data = pd.DataFrame()

    @abstractmethod
    def add_data(self, data_dict):
        pass

    def display_data(self):
        print(self.data)

class ConcretePandasPublisher(AbstractPandasPublisher):
    def add_data(self, data_dict):
        if not isinstance(data_dict, dict):
            raise ValueError("Input data should be a dictionary.")
        new_data = pd.DataFrame(data_dict)
        self.data = pd.concat([self.data, new_data], ignore_index=True)
```

```python
concrete_publisher = ConcretePandasPublisher()
data = {'Nome': ['Alice', 'Bob'], 'Idade': [25, 30]}
concrete_publisher.add_data(data)
concrete_publisher.display_data()
```

```python
import pandas as pd
from IPython.display import display

class PandasBase:
    def __init__(self):
        self.data = pd.DataFrame()

    def add_data(self, data_dict):
        try:
            if not isinstance(data_dict, dict):
                raise ValueError("Input data should be a dictionary.")
            new_data = pd.DataFrame(data_dict)
            self.data = pd.concat([self.data, new_data], ignore_index=True)
        except ValueError as ve:
            print(f"Error: {ve}")

    def display_data(self):
        display(self.data)

class PandasSubclass(PandasBase):
    def __init__(self, subclass_name):
        super().__init__()
        self.subclass_name = subclass_name

    def add_subclass_data(self, data_dict):
        try:
            if not isinstance(data_dict, dict):
                raise ValueError("Input data should be a dictionary.")
            new_data = pd.DataFrame(data_dict)
            new_data.columns = [f'{self.subclass_name}_{col}' for col in new_data.columns]
            self.data = pd.concat([self.data, new_data], axis=1)
        except ValueError as ve:
            print(f"Error: {ve}")
```

```python
# Exemplo de uso:
base_publisher = PandasBase()
base_data = {'Nome': ['Alice', 'Bob'], 'Idade': [25, 30]}
base_publisher.add_data(base_data)
print("Base Publisher Data:")
base_publisher.display_data()

subclass_publisher = PandasSubclass("Subclass")
subclass_data = {'Salário': [50000, 60000]}
subclass_publisher.add_subclass_data(subclass_data)
print("\nSubclass Publisher Data:")
subclass_publisher.display_data()
```
